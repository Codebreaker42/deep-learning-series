# creating and or & xor dataframe

```python
import pandas as pd
import seaborn as sns
import numpy as np
from mlxtend.plotting import plot_decision_regions
```

```python
and_data=pd.DataFrame()
or_data=pd.DataFrame()
xor_data=pd.DataFrame()
```

```python
print(and_data)
print(or_data)
print(xor_data)
```

```python
and_data['x1']=[0,0,1,1]
and_data['x2']=[0,1,0,1]
and_data['y']=[0,0,0,1]
```

```python
or_data['x1']=[0,0,1,1]
or_data['x2']=[0,1,0,1]
or_data['y']=[0,1,1,1]
```

```python
xor_data['x1']=[0,0,1,1]
xor_data['x2']=[0,1,0,1]
xor_data['y']=[0,1,1,0]
```

```python
# and data 
sns.scatterplot(x=and_data['x1'],y=and_data['x2'],hue=and_data['y'])
```

```python
# or data 
sns.scatterplot(x=or_data['x1'],y=or_data['x2'],hue=or_data['y'])
```

```python
# xor  data 
sns.scatterplot(x=xor_data['x1'],y=xor_data['x2'],hue=xor_data['y'])
```

```python
from sklearn.linear_model import Perceptron
clf1=Perceptron()
clf2=Perceptron()
clf3=Perceptron()
```

```python
clf1.fit(and_data.iloc[:,0:2].values,and_data.iloc[: ,-1].values)
clf2.fit(or_data.iloc[:,0:2].values,or_data.iloc[: ,-1].values)
clf3.fit(xor_data.iloc[:,0:2].values,xor_data.iloc[: ,-1].values)
```

```python
x=np.linspace(-1,1,5)
y=-x+1
```

```python
plt.plot(x,y)
sns.scatterplot(x=and_data['x1'],y=and_data['x2'],hue=and_data['y'],s=100)
```

```python
plot_decision_regions(and_data.iloc[:,0:2].values,and_data.iloc[:,-1].values, clf=clf1, legend=3)
```
