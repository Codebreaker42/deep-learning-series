```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('ushape.csv')
```

```python
df.head()
```

```python
df.rename(columns={'3.159499363321345566e-02':'xone','9.869877579082642072e-01':'xtwo','0.000000000000000000e+00':'y'},inplace=True)
```

```python
df.head()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
sns.scatterplot(x=df['xone'],y=df['xtwo'],hue=df['y'])
```

model without xavier and he wt initialization technique

```python
model1=Sequential()
model1.add(Dense(256,activation='relu',input_dim=2))
model1.add(Dense(128,activation='relu'))
model1.add(Dense(1,activation='sigmoid'))
model1.summary()
```

```python
model1.get_weights()
```

```python
model1.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history1=model1.fit(df.iloc[:,0:2].values,df.iloc[:,-1].values,validation_split=.1,epochs=100)
```

```python
plt.plot(history1.history['accuracy'])
plt.plot(history1.history['val_accuracy'])
```

```python
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
```

# model with xavier and he weight initialization 

use glorat/xavier when: activaton is sigmoid
,use he when : activation is relu

there are two types of weights initialization technique in both method 

1- uniform weight initialiazation 

2- normal weight initialization 

```python
from tensorflow.keras import initializers
```

```python
glorat=initializers.GlorotNormal()
he=initializers.HeNormal()
```

```python
model2=Sequential()
model2.add(Dense(256,activation='relu',kernel_initializer=he, input_dim=2))
model2.add(Dense(128,activation='relu',kernel_initializer=he))
model2.add(Dense(1,activation='sigmoid',kernel_initializer=glorat))
model2.summary()
```

```python
model2.get_weights()
```

```python
model2.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history2=model2.fit(df.iloc[:,0:2],df.iloc[:,-1],validation_split=.1,epochs=100)
```

```python
plt.plot(history2.history['accuracy'])
plt.plot(history2.history['val_accuracy'])
```

```python
plt.plot(history2.history['loss'])
plt.plot(history2.history['val_loss'])
```

decision boundary of both model

```python
from mlxtend.plotting import plot_decision_regions
```

```python
x=df.iloc[: ,0:2].values
y=df.iloc[:,-1].values
```

```python
figure=plt.figure(figsize=(14,4))
plt.subplot(121)
plt.suptitle('Model before and after weight initialization')
plot_decision_regions( x, y.astype('int'), clf=model1,legend=2)
plt.title('model before ')
plt.subplot(122)
plot_decision_regions(x,y.astype('int'),clf=model2,legend=2)
plt.title('model after')
```
